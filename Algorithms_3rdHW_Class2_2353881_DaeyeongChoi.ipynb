```python
from collections import deque

graph = {
    1: [2, 3, 4, 6],
    2: [1, 3],
    3: [2, 1, 5],
    4: [1, 6],
    5: [3, 6],
    6: [1, 4, 5]
}

distance = [-1] * 7
def bfs(start):
    queue = deque([start])
    distance[start] = 0
    while queue:
        v = queue.popleft()
        for i in graph[v]:
            if distance[i] == -1:
                distance[i] = distance[v] + 1
                queue.append(i)

bfs(1)
for i in range(1, 7):
    print(distance[i], end=' ')
# 출력: 0 1 1 1 2 1
```
