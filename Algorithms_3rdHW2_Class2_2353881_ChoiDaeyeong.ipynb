```python
nodes = ["점화", "계란 풀기", "수프 붓기", "라면 넣기", "라면 봉지 뜯기", "냄비에 물 붓기"]
name2idx = {name: idx for idx, name in enumerate(nodes)}

graph2 = {
    name2idx["점화"]: [name2idx["냄비에 물 붓기"], name2idx["라면 넣기"], name2idx["계란 풀기"]],
    name2idx["냄비에 물 붓기"]: [name2idx["수프 붓기"], name2idx["라면 봉지 뜯기"]],
    name2idx["라면 넣기"]: [],
    name2idx["계란 풀기"]: [],
    name2idx["라면 봉지 뜯기"]: [name2idx["라면 넣기"]],
    name2idx["수프 붓기"]: [name2idx["라면 넣기"]],
}

visited = [False]*len(nodes)
result = []

def dfs(v):
    visited[v] = True
    for i in graph2[v]:
        if not visited[i]:
            dfs(i)
    result.append(v)

for i in range(len(nodes)):
    if not visited[i]:
        dfs(i)

topo_order = [nodes[v] for v in reversed(result)]
print("위상 정렬 결과:", topo_order)
# 출력 예시: 위상 정렬 결과: ['점화', '냄비에 물 붓기', '라면 봉지 뜯기', '수프 붓기', '계란 풀기', '라면 넣기']
```
